```python
import csv
def readFiles(uniFile=r"C:\Users\AKSHAY KUMAR\akshay's folder\flm ds genai\mini_project-1\universities.csv",capitalfile=r"C:\Users\AKSHAY KUMAR\akshay's folder\flm ds genai\mini_project-1\capitals.csv"):
    all_univs={}
    countries={}
    with open(uniFile,mode='r') as file1:
        reader=csv.DictReader(file1)
        for row in reader:
            code=row['Code']
            all_univs[code]={
            "institute name":row["Institution name"],
            "world rank":row["World Rank"],
            "country":row["Country"],
            "national rank":row["National Rank"],
            "degrees":row["Degrees Offered"],
            "average cost":row["Average Cost"],
            "score":row["Score"]
        }
    with open(capitalfile,mode='r') as file2:
        country_reader=csv.DictReader(file2)

        for row in country_reader:
            country=row["Country Name"]
            countries[country]={
                "capital":row["Capital"],
                "latitude":row["Latitude"],
                "longitude":row["Longitude"],
                "country":row["Country Code"],
                "continent":row["Continent"]
            }
    
    for code,details in all_univs.items():
        co_country=details["country"]
        if co_country in countries:
            details["continent"]=countries[co_country]["continent"]

    return all_univs,countries
            

    

all_univs,countries=readFiles()

print(all_univs)
print(countries)
```

```python
def findCountryByNames(countryName,countries):
    for name,details in countries.items():
        if name.lower()==countryName.lower():
            return name,details['capital'],details["continent"]
        
findCountryByNames("india",countries)
```

```python
def findcountrybyname(countryname,countries):
    for name,details in countries.items():
        if name.lower()==countryname.lower():
            return {name:details}
        

findcountrybyname("india",countries)
```

```python
def getDistinctCountries(all_univs):
    distict_countries=set()
    for key,value in all_univs.items():
        distict_countries.add(value["country"])
    return distict_countries

getDistinctCountries(all_univs)
```

```python
def getTopRank(countryname,all_univs):
    best_university=None
    highest_rank=float('inf')
    for code,details in all_univs.items():
        if details["country"].lower()==countryname.lower():
            rank=int(details["world rank"])
            if rank<highest_rank:
                highest_rank=rank
                best_university=details["institute name"]


    return (highest_rank,best_university)


getTopRank("Germany",all_univs)

```

```python
def getTopNatRank(countryName,allUnivs):
    top_nat_rank_university=None
    highest_nat_rank=float("inf")
    for key,value in all_univs.items():
        if value["country"].lower()==countryName.lower():

            nat_rank=int(value["national rank"])
            if nat_rank<highest_nat_rank:
                top_nat_rank_university=value["institute name"]
                highest_nat_rank=nat_rank

    return highest_nat_rank,top_nat_rank_university


getTopNatRank("Japan",all_univs)
```

```python
def getAvgScore(countryName,all_univs):
    sum_of_score=0
    count=0
    for code,details in all_univs.items():
        if countryName.lower()==details["country"].lower():
            score=float(details["score"])
            sum_of_score+=score
            count+=1
    return (sum_of_score/count) if count>0 else None

getAvgScore("United Kingdom",all_univs)
```

```python
def getRelativeScoreContinent(countryName,all_univs):
    
    continent=None
    for code,details in all_univs.items():
        if details["country"].lower()==countryName.lower():
            continent=details["continent"]
            break
    if continent is None:
        return None
        
    sum_score=0
    count=0
    for code,details in all_univs.items():
        if details["country"].lower()==countryName.lower():
            sum_score+=float(details["score"])
            count+=1
    
    if count==0:
        return None
    average=sum_score/count

    highest_score=0

    for code,details in all_univs.items():
        if details["continent"].lower()==continent.lower():
            score=float(details["score"])

            if score>highest_score:
                highest_score=score

    return average/highest_score




getRelativeScoreContinent("United Kingdom",all_univs)
```

```python
def getUnivWithCapital(countryName,all_univs):
    codes=[]
    for code,details in all_univs.items():
        if details["country"].lower()==countryName.lower():
            codes.append(code)
    
    return codes if codes else None



getUnivWithCapital("Japan",all_univs)
```

```python
def getstudyInOnePlace(countryName,degrees,budget,all_univs):
    codes=[]
    for code,details in all_univs.items():
        if details["country"].lower()==countryName.lower() and all(degree in details["degrees"]for degree in degrees) and float(details["average cost"])<=budget:
            
            codes.append(code)

    return codes if codes else None

print(getstudyInOnePlace('Japan', ['Diploma', 'PhD'], 
25000,all_univs))
```

```python
def getstudyIntwoPlace(firstCode,firstDegree,secondCode,secondDegree,budget,allUnivs):
    if firstCode not in all_univs or secondCode not in all_univs:
        return False
    
    first_details=all_univs[firstCode]
    second_details=all_univs[secondCode]

    first_ok=(firstDegree in first_details["degrees"]
               and float(first_details['average cost'])<=budget
            )
    

    second_ok=(
        secondDegree in second_details["degrees"] 
        and float(first_details["average cost"])<=budget
    )


    return first_ok and second_ok


print(getstudyIntwoPlace('JDBS', 'Diploma', 'OTTE', 
'PhD', 40000,all_univs)) 

```
